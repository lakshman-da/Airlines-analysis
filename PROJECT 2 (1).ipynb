```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (already uploaded)
# file_path = "/mnt/data/68508af3-de84-4d3b-ad8a-9c7e35d67c2c.csv"
df = pd.read_csv("cl.csv")

# Set the style for the plots
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# 1. Overall Price Distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], kde=True, color="lightgreen")
plt.title("Overall Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# 2. Price Distribution by Class
plt.figure(figsize=(12, 6))
sns.boxplot(x='class', y='price', data=df, palette='Set2')
plt.title("Price Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Price")
plt.show()
```

# 2.Compare prices across airlines, stops, routes

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(x='airline', y='price', data=df, ax=axes[0])
axes[0].set_title("Price by Airline")
axes[0].tick_params(axis='x', rotation=90)

sns.boxplot(x='stops', y='price', data=df, ax=axes[1])
axes[1].set_title("Price by Stops")

top_routes = df['route'].value_counts().nlargest(5).index
sns.boxplot(x='route', y='price', data=df[df['route'].isin(top_routes)], ax=axes[2])
axes[2].set_title("Price by Top 5 Routes")
axes[2].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

# 3.Outlier detection (IQR/Z-score)

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = df[(df['price'] < Q1 - 1.5 * IQR) | (df['price'] > Q3 + 1.5 * IQR)]
print("Outliers using IQR method:", len(outliers_iqr))

df['z_price'] = zscore(df['price'])
outliers_z = df[df['z_price'].abs() > 3]
print("Outliers using Z-score method:", len(outliers_z))
```

# ### **Statistical Analysis**

- T-Test: Compare average fare between 0-stop and 1-stop
- ANOVA: Fare variation across multiple airlines
- Correlation between duration and price
- Chi-square: Association between total stops and price category

```python
from scipy import stats
import pandas as pd

# 1. T-Test: 0-stop vs 1-stop flights
df_ttest = df[df['stops'].isin(['zero', 'one'])].copy()
df_ttest['stops_num'] = df_ttest['stops'].map({'zero': 0, 'one': 1})
group_zero = df_ttest[df_ttest['stops_num'] == 0]['price']
group_one = df_ttest[df_ttest['stops_num'] == 1]['price']
ttest_stat, ttest_pval = stats.ttest_ind(group_zero, group_one, equal_var=False)

# 2. ANOVA: Price variation across airlines
anova_stat, anova_pval = stats.f_oneway(*[group['price'].values for _, group in df.groupby('airline')])

# 3. Correlation: Duration vs Price
correlation_coef = df['total_duration (minutes)'].corr(df['price'])
pearson_stat, pearson_pval = stats.pearsonr(df['total_duration (minutes)'], df['price'])

# 4. Chi-Square: Stops vs Price category
df['price_category'] = pd.qcut(df['price'], q=3, labels=["Low", "Medium", "High"])
df['stops_cat'] = df['stops'].map({'zero': '0', 'one': '1', 'two_or_more': '2+'}).fillna('Other')
contingency_table = pd.crosstab(df['stops_cat'], df['price_category'])
chi2_stat, chi2_pval, chi2_dof, chi2_expected = stats.chi2_contingency(contingency_table)

# Print all results
print("T-Test:", {"t-statistic": ttest_stat, "p-value": ttest_pval})
print("ANOVA:", {"F-statistic": anova_stat, "p-value": anova_pval})
print("Correlation:", {"coefficient": correlation_coef, "p-value": pearson_pval})
print("Chi-Square:", {"chi2-statistic": chi2_stat, "p-value": chi2_pval, "df": chi2_dof})
```

# Price by Airline

```python
airline_avg = df.groupby("airline")["price"].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(x=airline_avg.index, y=airline_avg.values, palette="viridis")
plt.title("Average Fare by Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Stops vs. Fare

```python
stops_avg = df.groupby("stops")["price"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x="stops", y="price", data=stops_avg, marker="o", linewidth=2)
plt.title("Stops vs. Average Fare")
plt.ylabel("Average Price")
plt.xlabel("Number of Stops")
plt.tight_layout()
plt.show()
```

# Source-Destination Routes

```python
route_avg = df.groupby("route")["price"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(x=route_avg.index, y=route_avg.values, palette="coolwarm")
plt.title("Top 10 Expensive Routes by Average Fare")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
